```python
# *** [Helper Functions] ***
import re 
    
# Function checks if the string contains any special character 
def stringHasSpecialChar(string): 

    # Make own character set and pass this as argument in compile method 
    regex = re.compile('[@_!#$%^&*()<>?/\|}{~:+=-`,;]') 

    # Pass the string in search method of regex object.     
    if regex.search(string) == None: 
        #print("String is accepted") 
        return False
    else: 
        #print("String is not accepted.")
        return True
    
# word = "yessir"
# print(stringHasSpecialChar(word))
```

```python
# *** [PART 1] ***
# ! The engineer explains that an engine part seems to be missing from the engine, but nobody can figure out which one. If you can add up all the *part numbers* in the engine schematic, it should be easy to work out which part is missing.
# - The engine schematic (your puzzle input) consists of a visual representation of the engine. 
# - There are lots of numbers and symbols that you don't really understand, but apparently any number *adjacent* 
#   to a symbol (even diagonally) = a "part number" and should be included in your sum. 
# - NOTE: Periods (.) do NOT count as a symbol.
# --------------------------------------------------------------------------------------------------------------------
# ! TODO determine if EACH number == a *part* number:
# - Standard Directionals: U = -1r; D = +1r; L = -1c; R = +1c
# - Diagonals: UL = -1r -1c, DL = +1r -1c; UR = -1r + 1c, DR = +1r +1c
# - Check U, D, L, R, & diagonals (UR, UL; DR, DL) of EACH number in the engime schematic map
# - Diagonals for numbers with more than 1 digit: L = [index of *first* digit - 1c]; R = [index of *last* digit + 1c]; 
#                                                 U = [index of *any* digit - 1r]; D = [index of *any* digit + 1r];
#                                                 UL = [index of *first* digit - 1r - 1c]; 
#                                                 UR = [index of *first* digit - 1r + 1c];
#                                                 DL = [index of *last* digit - 1r - 1c]; 
#                                                 DR = [index of *last* digit - 1r + 1c];
# - TODO for EACH number:
#   > Perform 5 checks on the *first* digit (U, L, D, UL, DL)
#   > Perform 5 checks on the *last* digit (U, R, D, UR, DR)
#   > NOTE: For single digit numbers: Perform 8 checks on the digit (U, L, D, UL, DL; R, UR, DR)
#   > NOTE: For numbers containing more than 2 digits: Perform additional U & D checks for EACH digit between 
#           the *first* and *last* digit
#--------------------------------------------------------------------------------------------------------------------
# TODO: For EACH line, find all numbers in the current line > store found numbers in an array list
# > For EACH number in the array list, store the length of the current number in a var
# > Find the index of the *first* digit of the current number (first occurrence of current number) in the current line
# > Start checks from the *first* digit (stored index) of the current number found in the current line up until the last digit 
#   of the current number
# > Index of last digit of current number = stored index of first digit of current number + (len(Num) - 1)

# TODO: Traverse through EACH line in the array list var 'arrFileData'
counter = 0
sumPartNumbers = 0

for line in arrFileData:
    counter = counter + 1 # Increment line counter
    sumLinePartNumbers = 0 

    arrNumbers = [] # Define array var to store list of numbers found in current line
    #numbers = line.split('.')
    lineSplit = re.split('[.!@#$%^&*]', line)
    #print(lineSplit)

    for value in lineSplit:
        # IF the current value in the list of split values of the current line == a number, then add the value to the
        # 'arrNumbers' var
        if value.isnumeric():
            arrNumbers.append(value)

    #print(arrNumbers)
    # TODO: Traverse through EACH number found in the current line
    if not(arrNumbers == []):
        for number in arrNumbers:
            # Store the length of the string representing the current number
            lenNum = len(number)

            firstDigitIdx = 0 # Stores the index of the first digit of the current number
            
            # TODO: Find the index of the first digit of the current number in the current line
            # - Traverse through the current string line, then find the FIRST occurrence of the current number in the 
            #   current line's string and store the index of the first digit of the current number found
            for i, char in enumerate(line):
                # IF number(E.g.'345') exists anywhere between the ith and 'i + len(number)'th indexed char 
                # of the current string line 
                if number in line[i:(i + lenNum)]: 
                    firstDigitIdx = i
                        
            # TODO: Apply checks to determine if current number = a part number
            Up = Down = Left = Right = False
            UR = UL = DR = DL = False

            # TODO: !!! HERE !!!
            # TODO: EDGE CASE #1: IF the current line == the *first* line in the 'arrFileData' array list
            # NOTE: No 'Up' checks => (U, UL, UR)
            if counter > 1:
                # TODO: Case for [single digit number] #1.1
                if lenNum == 1:
                    return True
                # TODO: Case for [number containing 2 digits] #1.2
                if lenNum == 2:
                    return True
                # TODO: Case for [number containing more than 2 digits] #1.3
                if lenNum > 2:
                    return True
                
            # TODO: EDGE CASE #2: IF the current line == the *last* line in the 'arrFileData' array list
            # NOTE: No 'Down' checks => (D, DL, DR)
            if counter < len(arrFileData): 
                # TODO: Case for [single digit number] #2.1
                if lenNum == 1:
                    return True
                # TODO: Case for [number containing 2 digits] #2.2
                if lenNum == 2:
                    return True
                # TODO: Case for [number containing more than 2 digits] #2.3
                if lenNum > 2:
                    return True

            # TODO: EDGE CASE #3: IF the current line is NEITHER the *first* or *last* line in the 'arrFileData' array list
            if (counter > 1) and (counter < len(arrFileData)):
                # TODO: Case for [single digit number] #3.1
                # NOTE: Checks => (U, L, D, UL, DL; R, UR, DR)
                if lenNum == 1:
                    Up = stringHasSpecialChar(arrFileData[counter - 1][firstDigitIdx])
                    
                    # TODO: Edge case check if the index of the *first* digit of the current number is NOT == 
                    # the index of the FIRST character of the current line = 0, 
                    # ELSE cannot perform 'Left' check, therefore 'Left = False'
                    if not(firstDigitIdx == 0):
                        Left = stringHasSpecialChar(arrFileData[counter][firstDigitIdx - 1]) # or line[firstDigitIdx - 1]??
                    
                    Down = stringHasSpecialChar(arrFileData[counter + 1][firstDigitIdx])
                    UL = stringHasSpecialChar(arrFileData[counter - 1][firstDigitIdx - 1])
                    DL = stringHasSpecialChar(arrFileData[counter + 1][firstDigitIdx - 1])

                    # TODO: Edge case check if the index of the *last* digit of the current number is NOT == 
                    # the index of the LAST character of the current line = len(line), 
                    # ELSE cannot perform 'Right' check, therefore 'Right = False'
                    if not(firstDigitIdx == len(line)): 
                        Right = stringHasSpecialChar(arrFileData[counter][firstDigitIdx + 1]) # or line[firstDigitIdx + 1]?? 
                    
                    UR = stringHasSpecialChar(arrFileData[counter - 1][firstDigitIdx + 1])
                    DR = stringHasSpecialChar(arrFileData[counter + 1][firstDigitIdx + 1])

                    # TODO: Check if the current number has a symbol (special char) adjacent to it
                    if (Up == True) or (Left == True) or (Down == True) or (UL == True) or (DL == True) or (Right == True) or (UR == True) or (DR == True):
                        sumLinePartNumbers = sumLinePartNumbers + int(number)

                # TODO: Case for [number containing 2 digits] #3.2
                # - Perform 5 checks on the *first* digit (U, L, D, UL, DL)
                # - Perform 5 checks on the *last* digit (U, R, D, UR, DR)
                if lenNum == 2:
                    secondDigitIdx = firstDigitIdx + 1

                    # TODO: FIRST digit checks
                    Up = stringHasSpecialChar(arrFileData[counter - 1][firstDigitIdx])

                    # TODO: Edge case check if the index of the *first* digit of the current number is NOT == 
                    # the index of the FIRST character of the current line = 0, 
                    # ELSE cannot perform 'Left' check, therefore 'Left = False'
                    if not(firstDigitIdx == 0):
                        Left = stringHasSpecialChar(arrFileData[counter][firstDigitIdx - 1]) # or line[firstDigitIdx - 1]??
                        
                    Down = stringHasSpecialChar(arrFileData[counter + 1][firstDigitIdx])
                    UL = stringHasSpecialChar(arrFileData[counter - 1][firstDigitIdx - 1])
                    DL = stringHasSpecialChar(arrFileData[counter + 1][firstDigitIdx - 1])

                    # TODO: LAST digit checks
                    Up = stringHasSpecialChar(arrFileData[counter - 1][secondDigitIdx])

                    # TODO: Edge case check if the index of the *last* digit of the current number is NOT == 
                    # the index of the LAST character of the current line = len(line), 
                    # ELSE cannot perform 'Right' check, therefore 'Right = False'
                    if not(firstDigitIdx == len(line)): 
                        Right = stringHasSpecialChar(arrFileData[counter][secondDigitIdx + 1]) # or line[secondDigitIdx + 1]?? 
                    
                    Down = stringHasSpecialChar(arrFileData[counter + 1][secondDigitIdx])
                    UR = stringHasSpecialChar(arrFileData[counter - 1][secondDigitIdx + 1])
                    DR = stringHasSpecialChar(arrFileData[counter + 1][secondDigitIdx + 1])

                    # TODO: Check if the current number has a symbol (special char) adjacent to it
                    if (Up == True) or (Left == True) or (Down == True) or (UL == True) or (DL == True) or (Right == True) or (UR == True) or (DR == True):
                        sumLinePartNumbers = sumLinePartNumbers + int(number)

                # TODO: Case for [number containing more than 2 digits] #3.3
                # - Perform 5 checks on the *first* digit (U, L, D, UL, DL)
                # - Perform 5 checks on the *last* digit (U, R, D, UR, DR)
                # - Perform additional U & D checks for EACH digit between the *first* and *last* digit
                if lenNum > 2:
                    lastDigitIdx = firstDigitIdx + (lenNum - 1)

                    # TODO: FIRST digit checks
                    Up = stringHasSpecialChar(arrFileData[counter - 1][firstDigitIdx])

                    # TODO: Edge case check if the index of the *first* digit of the current number is NOT == 
                    # the index of the FIRST character of the current line = 0, 
                    # ELSE cannot perform 'Left' check, therefore 'Left = False'
                    if not(firstDigitIdx == 0):
                        Left = stringHasSpecialChar(arrFileData[counter][firstDigitIdx - 1]) # or line[firstDigitIdx - 1]??
                    
                    Down = stringHasSpecialChar(arrFileData[counter + 1][firstDigitIdx])
                    UL = stringHasSpecialChar(arrFileData[counter - 1][firstDigitIdx - 1])
                    DL = stringHasSpecialChar(arrFileData[counter + 1][firstDigitIdx - 1])
                            
                    # TODO: MIDDLE digit checks
                    # - Determine how many digits exist inbetween the *first* and *last* digit of the current number
                    numMiddleDigits = lenNum - 2

                    # TODO: Perform U & D checks on the next 'numMiddleDigits' digits after the first digit 
                    # of the current number
                    for j in range(numMiddleDigits):
                        firstDigitIdx = firstDigitIdx + 1 # Move to check the next digit
                        Up = stringHasSpecialChar(arrFileData[counter - 1][firstDigitIdx])
                        Down = stringHasSpecialChar(arrFileData[counter + 1][firstDigitIdx])

                    # TODO: LAST digit checks
                    Up = stringHasSpecialChar(arrFileData[counter - 1][lastDigitIdx])

                    # TODO: Edge case check if the index of the *last* digit of the current number is NOT == 
                    # the index of the LAST character of the current line = len(line), 
                    # ELSE cannot perform 'Right' check, therefore 'Right = False'
                    if not(firstDigitIdx == len(line)): 
                        Right = stringHasSpecialChar(arrFileData[counter][lastDigitIdx + 1]) # or line[secondDigitIdx + 1]?? 
                    
                    Down = stringHasSpecialChar(arrFileData[counter + 1][lastDigitIdx])
                    UR = stringHasSpecialChar(arrFileData[counter - 1][lastDigitIdx + 1])
                    DR = stringHasSpecialChar(arrFileData[counter + 1][lastDigitIdx + 1])

                    # TODO: Check if the current number has a symbol (special char) adjacent to it
                    if (Up == True) or (Left == True) or (Down == True) or (UL == True) or (DL == True) or (Right == True) or (UR == True) or (DR == True):
                        sumLinePartNumbers = sumLinePartNumbers + int(number)

    sumPartNumbers = sumPartNumbers + sumLinePartNumbers
```
